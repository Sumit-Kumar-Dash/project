```python
data=pd.merge(new_meta,credits,on='id')
```

```python
pd.set_option('display.max_colwidth',75)
data
```

```python
# evaluates an expression node or a string containing a Python literal or container display
#converting the string to list
import ast
data['genres']=data['genres'].map(lambda x: ast.literal_eval(x))
data['cast']=data['cast'].map(lambda x: ast.literal_eval(x))
data['crew']=data['crew'].map(lambda x: ast.literal_eval(x) )
```

```python
data['genres']
```

```python
def gener_list(x):
    l=[]
    s=' '
    for i in x:
        if i.get('name')== 'Science Fiction':
            l.append("Sci-Fi")
        else:
            l.append(i.get('name'))
    if l==[]:
        return numpy.nan
    else:
        return s.join(l)
```

```python
data['genres_list']=data['genres'].map(lambda x: gener_list(x))
```

```python
data['genres_list']
```

```python
data['cast']
```

```python
def get_actor1(x):
    casts = []
    for i in x:
        casts.append(i.get('name'))
    if casts == []:
        return numpy.NaN
    else:
        return (casts[0])
```

```python
data['actor_1_name'] = data['cast'].map(lambda x: get_actor1(x))
```

```python
data['actor_1_name']
```

```python
for i in data['cast']:
    for j in i:
        print(j.get('name'))
    print('**********')
```

```python
def get_actor2(x):
    casts = []
    for i in x:
        casts.append(i.get('name'))
    if casts == [] or len(casts)<=1:
        return numpy.NaN
    else:
        return (casts[1])
```

```python
data['actor_2_name'] = data['cast'].map(lambda x: get_actor2(x))
```

```python
def get_actor3(x):
    casts = []
    for i in x:
        casts.append(i.get('name'))
    if casts == [] or len(casts)<=2:
        return numpy.NaN
    else:
        return (casts[2])
```

```python
data['actor_3_name'] = data['cast'].map(lambda x: get_actor3(x))
```

```python
def get_director(x):
    l=[]
    s=' '
    for i in x:
        if i.get('job')=='Director':
            l.append(i.get('name'))
    if l==[]:
        return numpy.nan
    else:
        return s.join(l)
```

```python
data['director_name']=data['crew'].map(lambda x:get_director(x))
```

```python
data['director_name']
```

```python
movie=data[['director_name','actor_1_name','actor_2_name','actor_3_name','genres_list','title']]
movie.head(2)
```
